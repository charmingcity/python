```python
# (1) 시리즈(series) : 1차원 배열 
# (2) 데이터프레임(dataframe) : 2차원 배열 (테이블 형태 )
```

```python
import pandas as pd
import numpy as np
```

```python
list_data = [1,2,3]
sr = pd.Series(list_data)
print(type(sr))
print(sr)
```

```python
list_data = [[1,2,3],
             [4,5,6],
             [7,8,9]]
df = pd.DataFrame(list_data)
print(type(df))
print(df)
```

```python
score_table = { '성명': ['조철민','김수연','백시열','김형석','박순재'],
                '국어': [50.,60,70,80,90],
                '영어': [10,20,30,40,50],
                '수학': [70,80,90,30,20]}

df = pd.DataFrame(score_table)
print(type(df))
df
```
