## 코스피. 코스닥 종목 정보 수집

```python
# 종목 목록 가져오기
stock_list = fdr.StockListing('KRX') 
```

```python
# 코넥스 종목 제외
stock_list = stock_list.loc[stock_list['Market'] != "KONEX"]
```

```python
# 지역이 있는 종목만 필터링하기
stock_list = stock_list.loc[stock_list['Region'].notnull()] # 선물, 우선주 등은 지역 정보가 없음
```

```python
# 종목 정보 내보내기 
stock_list.to_excel("stock_info.xlsx", index=False) 
```

```python
# 전 종목 순회 및 데이터 저장
import time
from tqdm import tqdm
for code, name in tqdm(stock_list[['Symbol', "Name"]].values):  #stock_list 데이터프레임에서 symbol과 name 칼럼을 각각 code와 name으로 동시에 순회 
    # print(code, name)
    while True:
        try:
            data = fdr.DataReader(code, "2011-01-01", "2022-06-10")
            if len(data) > 300:                   #데이터 길이가 300 이상으로 한정 
                data.to_excel("{}.xlsx".format(name))
            time.sleep(1)
            break # 정상적으로 데이터 저장까지 완료되면 반복문에서 빠져나감
        except:
            time.sleep(10 * 60) # 연결이 끊어지면 10분 재움
```
